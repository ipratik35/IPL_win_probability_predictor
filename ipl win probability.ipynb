## Finding the Current score, runsleft, ballsleft

```python
delivery_df['current_score'] = delivery_df.groupby('id')['total_runs_y'].cumsum()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

## To find how many wikcets left

```python
delivery_df.head()
```

```python
delivery_df['player_out'] = delivery_df['player_out'].fillna("0")
```

```python
delivery_df['player_out'] = delivery_df['player_out'].apply(lambda x:x if x == "0" else"1" )
```

```python
delivery_df['player_out'] = delivery_df['player_out'].astype('int') 
wickets = delivery_df.groupby('id')['player_out'].cumsum()

delivery_df['wickets'] = 10 - wickets
```

```python
delivery_df.tail()
```

## for current run rate = runs/ overs

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df
```

## For required run rate  = (runsleft*6) /balls left

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df.head(2)
```

## to convert winning team result into 0 or 1 

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr', 'result' ]]
```

```python
final_df
```

## Shuffling the dataset for not getting baised results

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

## Dropping the column where values are nan

```python
final_df.dropna(inplace= True)
```

## Now remove the rows where balls left = 0 because due to this rows here required run rate is showing infinite
## i.e. when balls are 0 and still team requires to score run then it shows infinite rrr

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
final_df.isnull().sum()
```

## Model building

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:, :-1]
y = final_df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC



```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')

```

```python
pipe = Pipeline(steps=[
    ('step1', trf),     # StandardScaler as the first step
    ('step2', LogisticRegression(solver= 'liblinear')),
    
    
    
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

## Here when we use the randomforest it gives 99% which is overfitiing but logistic regression gives best fit results
